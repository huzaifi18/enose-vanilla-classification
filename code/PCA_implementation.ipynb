```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
df_1w = pd.read_csv("../dataset/data_2c40d1w.csv")
df_2w = pd.read_csv("../dataset/data_2c40d2w.csv")
df_3w = pd.read_csv("../dataset/data_2c40d3w.csv")
```

```python
X_1w = df_1w.drop(columns="label").values
X_2w = df_2w.drop(columns="label").values
X_3w = df_3w.drop(columns="label").values
```

```python
scaler_1w = StandardScaler()
scaler_2w = StandardScaler()
scaler_3w = StandardScaler()
```

```python
X_1w_standardized = scaler_1w.fit_transform(X_1w)
X_2w_standardized = scaler_2w.fit_transform(X_2w)
X_3w_standardized = scaler_3w.fit_transform(X_3w)
```

```python
y_1w = df_1w.label.to_numpy()
y_2w = df_2w.label.to_numpy()
y_3w = df_3w.label.to_numpy()
```

```python
pca_1w = PCA(n_components=2)
X_1w_pca = pca_1w.fit_transform(X_1w_standardized)

pca_2w = PCA(n_components=2, whiten=True)
X_2w_pca = pca_2w.fit_transform(X_2w_standardized)

pca_3w = PCA(n_components=2, whiten=True)
X_3w_pca = pca_3w.fit_transform(X_3w_standardized)
```

```python
sns.scatterplot(x=X_1w_pca[:,0], y=X_1w_pca[:,1], hue=y_1w);
plt.title('PCA of 1 Window Dataset')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
sns.scatterplot(x=X_2w_pca[:,0], y=X_2w_pca[:,1], hue=y_2w);
plt.title('PCA of 2 Window Dataset')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
sns.scatterplot(x=X_3w_pca[:,0], y=X_3w_pca[:,1], hue=y_3w);
plt.title('PCA of 3 Window Dataset')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```
