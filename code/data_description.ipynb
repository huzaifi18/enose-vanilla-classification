## Menyiapkan semua library yang dibuthkan

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from baseline_correction import normalize_data_with_Baseline
```

## Data Wrangling

### Gathering Data

```python
A_df = pd.read_csv("../data_raw/A_1.csv")
B_df = pd.read_csv("../data_raw/B_1.csv")
```

# Sampel Data Visualization

## Label A

```python
att_names = ['F1', 'F2', 'F3', 'F4', 'F5', 'F6', 'F7', 'F8', 'F9', 'F10']
```

```python
fig, ax=plt.subplots(figsize=(15, 7))
for i in att_names:
    plt.plot(A_df['time(s)'], A_df[i], label=i, linewidth=3)

plt.xlabel('Time (second)', fontsize=18)
plt.ylabel('Value', fontsize=18) # ganti satuan
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=18)

# Display the plot
plt.show()
```

```python
A_df_crop = A_df[(A_df['time(s)'] >= 10) & (A_df['time(s)'] <= 30)]

fig, ax=plt.subplots(figsize=(15, 7))
for i in att_names:
    plt.plot(A_df_crop['time(s)'], A_df_crop[i], label=i, linewidth=3)

plt.xlabel('Time (second)', fontsize=18)
plt.ylabel('Value', fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=18)

# Display the plot
plt.show()
```
